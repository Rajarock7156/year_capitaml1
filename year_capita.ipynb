```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r'D:\ml\py-master\py-master\ML\1_linear_reg\Exercise\canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='*')


```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
reg.predict(np.array([[2023]]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='*')
plt.scatter(df.year,reg.predict(df[['year']]),marker='+')
```

```python
op=reg.predict(np.array([[2017],[2018],[2019],[2020],[2021],[2022],[2023],[2024]]))
op
```
